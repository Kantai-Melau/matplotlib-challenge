## Observations and Insights 
1. There is an almost equal distribution of the mice tested in terms of sex.
2. There are no outliers in the data.
3. There is a direct positive relationship between the weight of the mice and the average tumor volume.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_data = pd.merge(mouse_metadata, study_results, on = "Mouse ID", how = "outer")
# Display the data table for preview
merged_data.head(20)
```

```python
# Checking the number of mice.
mouse_count = merged_data.value_counts("Mouse ID")
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merged_data[merged_data.duplicated(["Mouse ID", "Timepoint"])]
duplicate_count = duplicate_mice["Mouse ID"].value_counts()
duplicate_count
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = merged_data.drop_duplicates(subset = ["Mouse ID", "Timepoint"])
clean_data
```

```python
# Checking the number of mice in the clean DataFrame
mice_count = clean_data["Mouse ID"].value_counts()
mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

group_data = clean_data.groupby(["Drug Regimen"])

mean_data = group_data["Tumor Volume (mm3)"].mean()
median_data = group_data["Tumor Volume (mm3)"].median()
variance_data = group_data["Tumor Volume (mm3)"].var()
std_data = group_data["Tumor Volume (mm3)"].std()
sem_data = group_data["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.

summary_df = pd.DataFrame()

summary_df["Mean"] = mean_data
summary_df["Median"] = median_data
summary_df["Variance"] = variance_data
summary_df["Standard Deviation"] = std_data
summary_df["SEM"] = sem_data
summary_df.head()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
target_data = group_data["Tumor Volume (mm3)"]
aggregate_data = target_data.aggregate(["mean", "median", "var", "std", "sem"])
aggregate_data
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(capomulin_mouse["Tumor Volume (mm3)"], capomulin_mouse["Weight (g)"])
print(f" The correlation coefficient between weight and tumor volume is {round(correlation[0],2)}")
```

```python
c_slope, c_int, c_r, c_p, c_std_err = st.linregress(capomulin_mouse["Tumor Volume (mm3)"], capomulin_mouse["Weight (g)"])
c_fit = c_slope * capomulin_mouse["Tumor Volume (mm3)"] + c_int

plt.scatter(capomulin_mouse["Tumor Volume (mm3)"], capomulin_mouse["Weight (g)"])
plt.plot(capomulin_mouse["Tumor Volume (mm3)"],c_fit,"--")
plt.xlabel("Tumor Volume (mm3)")
plt.ylabel("Weight (g)")
plt.show()
```
